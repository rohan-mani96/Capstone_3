# 5.3 Imports<a id='5.3_Imports'></a>

```python
# from sklearn import train_test_split
import pickle
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn import preprocessing
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import shap
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
shap.initjs()
```

# 5.4 Load Data<a id='5.4_Load_Data'></a>

```python
X_test=pd.read_csv('../data/processed/X_test.csv')
y_test=pd.read_csv('../data/processed/y_test.csv')
```

# 5.5 Load Model<a id='5.5_Load_Model'></a>

```python
model=pickle.load(open('../models/xgb','rb'))
# model2=pickle.load(open('../models/sklearn_gbm','rb'))
```

```python
model=pickle.load(open('../models/xgbv2','rb'))
```

# 5.6 Setup Pipeline to Encode/Process Test Data<a id='5.6_Setup_Pipeline_to_Encode/Process_Test_Data'></a>

```python
categorical_features=['state','city']

transformer = ColumnTransformer(transformers=[('c', preprocessing.OrdinalEncoder(), categorical_features)])

scaler=preprocessing.MinMaxScaler()

pipe = Pipeline(steps=[("encoder",transformer),("scaler", scaler), ("m", model)])
```

```python
X_test.isnull().sum()
```

# 5.7 Execute Pipeline and Generate Model Predictions<a id='5.7_Execute_Pipeline_and_Generate_Model_Predictions'></a>

```python
X_test[['state','city']]=pipe[0].fit_transform(X_test)
X_test=pd.DataFrame(pipe[1].fit_transform(X_test),columns=X_test.columns)
pipe[2].predict(X_test)
```

```python
y_test
```

```python
pred=pipe[2].predict(X_test)
```

```python
r2_score(pred,y_test['price'])
```

```python
X_test
```

```python
model.feature_importances_
```

## 5.8.1 Plot Actual vs Predictions<a id='5.8.1_Plot_Actual_vs_Predictions'></a>

```python
plt.scatter(y_test,pred,alpha=0.2)
a=y_test
p=y_test
plt.plot(a,p,c='r')
# Label it nicely
plt.xlabel('Actual house price')
plt.ylabel('Predicted house price')

plt.show()
```

```python
explainer=shap.TreeExplainer(model)
```

```python
shap_values=explainer.shap_values(X_test)
```
